```python
import folium as f 
import kml_parser_functs as kml
import map_functs as mf 
import math as m 
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon
import matrixmath as mm 
import pandas as pd
```

```python
filename = 'GoogleEarth_KML/Bodium_Vineyard.kml'
filename = 'GoogleEarth_KML/Gusbourne_Vineyard.kml'
```

```python
name,lookat, features = kml.parse_KML(filename)
```

```python
ax = plt.gca()
ax.cla()
for feature in features:
    x = []
    y = []
    for p in feature.matrix_coords: 
    
        x.append(p[0])    
        y.append(p[1])
           
    xr = np.array(x)
    yr = np.array(y)
    
    x = []
    y = []
    
    for coord in feature.coords:
        x.append(coord[1])
        y.append(coord[0])
        
    poly_x = np.array(x)
    poly_y = np.array(y)
    
    plt.plot(poly_x,poly_y, color = feature.color)
    plt.scatter(yr, xr, color = feature.color, marker = "1")

ax.set_aspect('equal', adjustable= 'box')
plt.show()
```
