```python
from os import walk
mypath = 'SampleData/Bodium_Vineyard/'
files = []
for (dirpath, dirnames, filenames) in walk(mypath):
    files.extend(filenames)
    break
```

```python
data = []
for file in files:
    fp = mypath+file
    df = pd.read_csv(fp)
    coords = df['Coord'].tolist()
    esca = df['ESCA'].tolist()
    humidity = df['Humidity'].tolist()
    op = []
    for i in range(len(coords)):
        x = [float(i) for i in coords[i].strip('()').split(',')]
        x.append(esca[i])
        op.append(x)

    data.append(op)
    
#HeatMap(op).add_to(f.FeatureGroup(name='Heat Map').add_to(m))


HeatMapWithTime(data).add_to(f.FeatureGroup(name='Heat Map').add_to(m))
f.LayerControl().add_to(m)
m
```
